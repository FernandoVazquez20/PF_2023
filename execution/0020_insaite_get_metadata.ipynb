```
import pandas as pd
```

```
def get_metadata_df(df_002: pd.DataFrame):
  """
  Description:

  Take a dataframe,return metadata and create a dataframe with dimension time.

  Argument: 

  df_002: an pd.DataFrame

  Returns:
  
  rows_size: an int variable.
  cols_size: an int variable.
  X: an pd.DataFrame.
  Y: an  pd.DataFrame.
  df_dim_time:  an pd.DataFrame. 



  """

  rows_size,cols_size = df_002.shape
  X =  df_002[['InvoiceNo','CustomerID','Country','InvoiceDate','UnitPrice']]
  Y = df_002[['Quantity','StockCode']]
  df_dim_time = pd.DataFrame()
  df_dim_time['Year'] =  df_002['InvoiceDate'].dt.year
  df_dim_time['Month'] =  df_002['InvoiceDate'].dt.month
  df_dim_time['Day'] =  df_002['InvoiceDate'].dt.day
  df_dim_time['Qtr'] =  df_002['InvoiceDate'].dt.quarter

  return rows_size,cols_size,X,Y,df_dim_time
```
