```
import pandas as pd
import pandera as pa


from pandera.dtypes import DateTime
```

```
def read_file_sales(archivo):

  """
  Description: 

  Validates the integrity of the df schema,
  if it meets the criteria, it returns a df for the next function.

  Argument:

  archivo: an pd.DataFrame 

  Returns:

  Try: an Schema error pandera type.

  Exception: an pd.DataFrame with the information that complies with the scheme.

  """

  df_001 = pd.read_excel(archivo)

  # --- EXAMPLE OF VALIDATION OF NUMERICAL DATA RANGE IN CASE IT IS NEEDED --- #
  df_qty_min,df_qty_max = df_qty_min,df_qty_max = df_001['Quantity'].min(),df_001['Quantity'].max()

  # --- Scheme to validate. --- #
  schema = pa.DataFrameSchema({
    "InvoiceNo": pa.Column(str,nullable=False),
    "StockCode": pa.Column(str,nullable=False),
    "Description": pa.Column(str,nullable=False),
    "Quantity": pa.Column(int,checks=pa.Check(lambda s: s.between(df_qty_min,df_qty_max)),nullable=False),
    "InvoiceDate": pa.Column(DateTime,nullable=False),
    "UnitPrice": pa.Column(float,nullable=False),
    "CustomerID": pa.Column(float,nullable=False),
    "Country": pa.Column(str,nullable=False)
  })

  try:
    schema(df_001,lazy=True)
    
    

  except pa.errors.SchemaErrors as exc:

    df_wiht_valid_data = df_001[~df_001.index.isin(exc.failure_cases["index"])]

  return  df_wiht_valid_data
```
